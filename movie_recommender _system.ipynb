```python
import ast
def convert(obj):
    ans = []
    for i in ast.literal_eval(obj):
        ans.append(i['name'])
    return ans    
```

```python
data['genres'] = data['genres'].apply(convert)
```

```python
data
```

```python
data.iloc[0]['keywords']
```

```python
data['keywords'] = data['keywords'].apply(convert)
```

```python
data
```

```python
def convert_cast(obj):
    l = []
    counter = 0;
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['character'])
            counter+=1
        else:
            break
    return l
```

```python
data['cast'] = data['cast'].apply(convert_cast)
```

```python
data
```

```python
data.iloc[0]['cast']
```

```python
data.iloc[0]['overview']
```

```python
data.iloc[0]['crew']
```

```python
def convert_crew(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
data['crew'] = data['crew'].apply(convert_crew)
```

```python
data.head()
```

```python
def convert_list(obj):
    try:
        return ast.literal_eval(obj)
    except (ValueError, SyntaxError):
        return obj  # Return the original object if it's not a valid literal

data['overview'] = data['overview'].apply(convert_list)
```

```python
data
```

```python
data['cast'] = data['cast'].apply(lambda x:[i.replace(" " ,"" ) for i in x])
data['keywords'] = data['keywords'].apply(lambda x:[i.replace(" " ,"" ) for i in x])
data['genres'] = data['genres'].apply(lambda x:[i.replace(" " ,"" ) for i in x])
```

```python
data['crew'] = data['crew'].apply(lambda x:[i.replace(" " ,"" ) for i in x])
```

```python
data
```

```python
# Convert all elements to strings and handle NaN values
data['overview'] = data['overview'].fillna('').astype(str)
data['genres'] = data['genres'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x)).fillna('')
data['keywords'] = data['keywords'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x)).fillna('')
data['cast'] = data['cast'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x)).fillna('')
data['crew'] = data['crew'].fillna('').astype(str)
```

```python
data
```

```python
data['tag'] = data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['crew']
```

```python
data
```

```python
newdata = data[['movie_id','title','tag']]
```

```python
newdata.iloc[0]
```

```python
newdata['tag'] = newdata['tag'].apply(lambda x:x.lower())
```

```python
newdata
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def convert_stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)        
```

```python
newdata['tag'] = newdata['tag'].apply(convert_stem)
```

```python
newdata
```

# Similarity Calculation

```python
from sklearn.feature_extraction.text import CountVectorizer
```

# Text to Vector

```python
# Initialize the CountVectorizer
vectorizer = CountVectorizer(max_features = 5000,stop_words=  'english')
# Fit and transform the text data to a bag of words model
vectors = vectorizer.fit_transform(newdata['tag']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
vectorizer.max_features
```

```python
print(vectorizer.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_metrics = cosine_similarity(vectors)
```

```python
similarity_metrics
```

```python
similarity_metrics[0]
```

# Main Function 

```python
def Movie_Recommender(movie_name):
    movie_index = newdata[newdata['title']==movie_name].index[0]
    # print(movie_index)
    similarity_row = similarity_metrics[movie_index]
    # print(similarity_row)
    
    enumerated_vector = list(enumerate(similarity_row))

    sorted_enumerated_vector = sorted(enumerated_vector,reverse = True, key=lambda x: x[1])

    # Extract the sorted values and their original indices
    sorted_values = [x[1] for x in sorted_enumerated_vector]
    # print(sorted_values)
    original_indices = [x[0] for x in sorted_enumerated_vector]
    recommender_index = original_indices[1:6]
    ans = []
    for i in recommender_index:
        ans.append(newdata.iloc[i]['title'])
    print(ans)
    
```

```python
Movie_Recommender('Newlyweds')
```
